```python
aa1 = df["Annual Income (k$)"][(df["Annual Income (k$)"]>=0) & (df["Annual Income (k$)"]<=30)]
aa2 = df["Annual Income (k$)"][(df["Annual Income (k$)"]>=31) & (df["Annual Income (k$)"]<=60)]
aa3 = df["Annual Income (k$)"][(df["Annual Income (k$)"]>=61) & (df["Annual Income (k$)"]<=90)]
aa4 = df["Annual Income (k$)"][(df["Annual Income (k$)"]>=91) & (df["Annual Income (k$)"]<=120)]
aa5 = df["Annual Income (k$)"][(df["Annual Income (k$)"]>=121) & (df["Annual Income (k$)"]<=150)]

aax = ["1-30","31-60","61-90","91-120","121-150"]
aay = [len(aa1.values),len(aa2.values),len(aa3.values),len(aa4.values),len(aa5.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=aax,y=aay,palette="Spectral")
plt.title("Annual Income (k$)")
plt.xlabel("Income")
plt.ylabel("Number of Customer")
plt.show()
```

```python
x1 = df.loc[:, ["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,4))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color='red',marker="8")
plt.xlabel("K Value")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(x1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x1[:,0],x1[:,1],c=kmeans.labels_,cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="black")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
